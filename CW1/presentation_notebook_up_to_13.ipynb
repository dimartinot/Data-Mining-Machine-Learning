### 5. [Deeper analysis of the data: the data is split into 10 classes, search for important features on each class]

For this intent, we have used the *KBEST* method from Scikit-Learn that calculates the k-highest score for each attribute to see how good it correlates with the class attribute.

But first, we need to fix the imbalances in our dataset where some classes, like the n°7, are in huge under representation in the case of the "*1-vs-all*" problem.
This is the distribution of each class in the "*1-vs-all*" training set:

```python
count_zeros_and_ones(y_train_one_vs_all)
```

To balance it, we will try using first an Oversampling method called [SMOTE](https://imbalanced-learn.readthedocs.io/en/stable/generated/imblearn.over_sampling.SMOTE.html). This algorithm will generate artificial positive elements to balance our dataset. As some classes like n°7 only have 167 positive elements, just oversampling will not be enough, we will also  perform downsampling over the negative instances of the newly generated dataset. This aims to have a better balance between precision and recall. 

We will do this in 2 steps:

 1) Run SMOTE to perform the oversampling on our categorical data (to have a 2/3 of 0s and 1/3 of ones balance)
 
 2) Run Random Under Sampling to perform undersampling (to have 1/2 of each)
 

```python
# we scale all 255 values to be at 1
scaled_X_train_dis_arr = X_train_dis_arr / 255

# we only aim for a 1/3 of imbalance data. We will then fix this to a 1/2 using undersampling
sm = imblearn.over_sampling.SMOTE(
    sampling_strategy = 1/3, 
    random_state=SEED)

new_train_sets = {}

for i in CLASSES.keys():
    new_train_sets[i], new_y_train_i  = sm.fit_sample(scaled_X_train_dis_arr, np.ravel(y_train_one_vs_all[i]))
    # we round every pixel value (between 0 and 1) and multiply it by 255 to retrieve an image with ONLY 0 and 255 pixels value
    new_train_sets[i] = 255*np.round(new_train_sets[i])
    y_train_one_vs_all[i] = pd.DataFrame(new_y_train_i)

count_zeros_and_ones(y_train_one_vs_all)
```

We can see above our new distribution.
We will now perform undersampling to minimize the number of 0 instances:

```python
us = imblearn.under_sampling.RandomUnderSampler(
    sampling_strategy = "majority", 
    random_state = SEED)

for i in CLASSES.keys():

    new_train_sets[i], new_y_train_i  = us.fit_sample(new_train_sets[i], np.ravel(y_train_one_vs_all[i]))
    y_train_one_vs_all[i] = pd.DataFrame(new_y_train_i)

count_zeros_and_ones(y_train_one_vs_all)
```

Now that our data is balanced, we can select the 20 most relevant attributes for each class.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

kbest_model_per_class = {}
best_attributes_per_class = {}

for i in CLASSES.keys():
    #k argument only relevant in case of the direct use of fit_transform
    kbest_model_per_class[i] = SelectKBest(score_func=chi2, k=100)
    kbest_model_per_class[i].fit(new_train_sets[i], y_train_one_vs_all[i])
    score = kbest_model_per_class[i].scores_
    
    # the 2*NUM_FIELDS_TO_KEEP specifies the number of attributes to keep (here 20)
    tmp_score_arr = sorted(zip(score, list(range(new_train_sets[i].shape[1]))), reverse=True)[:2*NUM_OF_FIELDS_TO_KEEP]
    best_attributes_per_class[i] = [index for (_, index) in tmp_score_arr]
```

We can now see which attributes are the best.

```python
for i in CLASSES.keys():
    print("Best class attributes for class n°{} is: \t{}".format(i, best_attributes_per_class[i]))
```

### 6. [Try to improve the classification, based on information from item 5]

However, just seeing the 20 best attributes did not seem to really guide us to ways to improve our model:
We have then decided to dive deeper in the relationship linking the number of attributes to the prediction performance by trying many different configuration of class attributes (from 10 to 2000 with a step of 10, for a total of 200 different datasets on which we will train 10 different models, 1 per class, resulting in a total of 2000 models trained).

We will now try to see how many kept attributes is optimal for our Naive Bayes algorithm by launching one vs all training on multiple of tens of attributes (10, 20, 30, 40... to 2000) for each class, creating 200 differently trained models for each class (total of 2000 models).

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

kbest_model_per_class = {}
list_of_best_attributes_per_class = {}


for i in CLASSES.keys():
    kbest_model_per_class[i] = SelectKBest(score_func=chi2, k=NUM_FIELDS)
    kbest_model_per_class[i].fit(new_train_sets[i], y_train_one_vs_all[i])
    score = kbest_model_per_class[i].scores_
    
    # we now loop over j to get the 10, 20, 30, 40 etc.. best attributes
    list_of_best_attributes_per_class[i] = []
    for j in range(1,NUM_OF_TENS_OF_FIELDS_TO_TRAIN+1):
    
        tmp_score_arr = sorted(zip(score, list(range(new_train_sets[i].shape[1]))), reverse=True)[:j*NUM_OF_FIELDS_TO_KEEP]
        list_of_best_attributes_per_class[i].append([index for (_, index) in tmp_score_arr])
```

We created here an automatic Bayesian Network that uses the 10 KBest pixels of our dataset. There exist different algorithms to generate the network. We will now compare how they influence the performance. We will, at the same time, explore how the number of kept pixels also helps.

```python
def net_predict(net, line):
    return net.predict([line])

algorithms = ['greedy', 'exact', 'chow-liu']

# stores array of performances when training on different size of pixels
performances = {algorithm: {"acc":[], "recall":[], "precision":[]} for algorithm in algorithms}

# takes around 1.5hour
for num_of_pixels in range(2, 20):
    
    kbest_filter = SelectKBest(chi2, k=num_of_pixels).fit(binary_X_train, y_train)
    binary_X_train_filtered = kbest_filter.transform(binary_X_train)
    binary_X_test_filtered = kbest_filter.transform(binary_X_test)
    binary_X_test_filtered_with_none = np.c_[binary_X_test_filtered, np.full((binary_X_test_filtered.shape[0], 1), None)]
    
    # for each training algorithm, we will develop one version of the graph
    for algorithm in algorithms:
            tmp_net = BayesianNetwork().from_samples(np.c_[binary_X_train_filtered, y_train.values], algorithm=algorithm) 
            
            #We need to encapsulate the line in an array so that its has this shape: [[...]]
            pred_line = lambda x: tmp_net.predict([x]) 
            
            preds = np.apply_along_axis(pred_line, axis=1, arr=binary_X_test_filtered_with_none)
            
            y_pred = preds[:,0, -1].reshape(-1, 1).astype(np.uint8)
                                    
            acc = metrics.accuracy_score(y_test, y_pred)
            precision = metrics.precision_score(y_test, y_pred, average="macro")
            recall = metrics.recall_score(y_test, y_pred, average="macro")
            
            performances[algorithm]["acc"].append(acc)
            performances[algorithm]["precision"].append(precision)
            performances[algorithm]["recall"].append(recall)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 8))

x = np.linspace(0, 17, num = 18)

for algorithm in algorithms:
    ax1.plot(x, performances[algorithm]["acc"], label=algorithm)
    ax2.plot(x, performances[algorithm]["precision"], label=algorithm)
    ax3.plot(x, performances[algorithm]["recall"], label=algorithm)

ax1.set_title("Accuracy of the 3 different Models")
ax1.legend(loc="upper right")
ax2.set_title("Precision of the 3 different Models")
ax2.legend(loc="upper right")
ax3.set_title("Recall of the 3 different Models")
ax3.legend(loc="upper right")

fig.show()
```

### 12. [Beyond  k-means,  tools  for  computation  of  optimal  number  of  clusters]

We will now try out different other clustering algorithms, explaining briefly their role and showing their results.

First, we use the Affinity Propagation method, it is generally used to work with many clusters of uneven size. It creates clusters by communicating between pairs of samples until convergence. We obtained a fair precision of 86%. 

```python
from sklearn.cluster import AffinityPropagation
global_AF = AffinityPropagation()
global_AF.fit(reduced_X_train)

pred_af = global_AF.predict(reduced_X_test)
accuracy, conf_matrix = score_kmean(y_test, pred_af)
sns.heatmap(conf_matrix)
```

Then we tried using the Gaussian Mixture probabilistic method, generally used for density estimations but is struggling with uneven data distribution, it scored a 78% accuracy 

```python
from sklearn import mixture
global_mixture = mixture.GaussianMixture(n_components=20, covariance_type='full')
global_mixture.fit(reduced_X_train)

pred_mixture = global_mixture.predict(reduced_X_test)
accuracy, conf_matrix = score_kmean(y_test, pred_mixture)
sns.heatmap(conf_matrix)
```

Then we tried using the Birch method that can be used in large dataset and calculate the Euclidean distance between points and creates a Tree of characteristic features of the data. By setting the number of clusters to NONE (the final clustering step is not performed, and the sub clusters are returned as they are. cf. Documentation) the precision goes up to 98%.  

```python
from sklearn.cluster import Birch
global_birch = Birch(n_clusters=None)
global_birch.fit(reduced_X_train)

pred_birch = global_birch.predict(reduced_X_test)
accuracy, conf_matrix = score_kmean(y_test, pred_birch)
sns.heatmap(conf_matrix)
```

We will now try using a different method that tries to identify blobs in a sample, the Mean Shift method. As our data is made of pixels, we do not expect this method to work well and in fact it scores a 18% accuracy by only predicting one specific class. 

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
bandwidth = estimate_bandwidth(X_train, quantile=0.2, n_samples=500)
global_shift = MeanShift()
global_shift.fit(reduced_X_train)

pred_shift = global_shift.predict(reduced_X_test)
accuracy, conf_matrix = score_kmean(y_test, pred_shift)
sns.heatmap(conf_matrix)
```

### 13. [Make conclusions]
Birch is by far the best method we can use and is the fastest to execute. We saw through this experimentation that the best methods are those that can work with many clusters with ease regarding of their size (Affinity propagator and Birch). On the other hand, density-based techniques perform worse than the others as they can’t assign a class to a cluster consistently (Gaussian Mixture and Mean Shift). 

As Birch exploits the observation that data space is not usually uniformly occupied and not every data point is equally important, it shows how some examples are less relevant than others, something that KMeans does not do. A solution to this would be to eventually weight examples when performing KMeans.
